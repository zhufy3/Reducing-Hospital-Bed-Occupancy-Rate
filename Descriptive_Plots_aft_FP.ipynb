```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train=pd.read_csv("clean_aft_fp.csv")
train = train.sort_values(by=["Stay","Age","Available Extra Rooms in Hospital","Bed Grade","Hospital_type_code","Hospital_region_code"])
```

### 1. Boxplot for numerical variables

```python
fig, ax = plt.subplots(figsize = (12,6))    
sns.boxplot(y = 'Stay', x="Available Extra Rooms in Hospital", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.boxplot(y = 'Stay', x="Hospital_code", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.boxplot(y = 'Stay', x="City_Code_Hospital", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.boxplot(y = 'Stay', x="Bed Grade", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.boxplot(y = 'Stay', x="Admission_Deposit", data = train, palette = 'Set2')
plt.show()
```

### 2. Heatmap for category variables

```python
train = train.sort_values(by=["Stay","Hospital_type_code"])

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Hospital_type_code", data = train, palette = 'Set2')
plt.show()

train = train.sort_values(by=["Stay","Hospital_region_code"])

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Hospital_region_code", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Department", data = train, palette = 'Set2')
plt.show()

train = train.sort_values(by=["Stay","Ward_Type"])


fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Ward_Type", data = train, palette = 'Set2')
plt.show()

train = train.sort_values(by=["Stay","Ward_Facility_Code"])


fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Ward_Facility_Code", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Type of Admission", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Age", data = train, palette = 'Set2')
plt.show()

fig, ax = plt.subplots(figsize = (12,6))    
sns.histplot(y = 'Stay', x="Severity of Illness", data = train, palette = 'Set2')
plt.show()
```
